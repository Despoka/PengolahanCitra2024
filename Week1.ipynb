```python
img = cv2.imread('assets/images/semangka.jpg')
print("DImension for image is{}".format(img.shape))

Length = img.shape[0]
width = img.shape[1]
Channels = img.shape[2]

print("Length of image is {}".format(Length))
print("Width of image is {}".format(width))
print("No of channels is {}".format(Channels))
```

```python
newImg = cv2.resize(img,(0,0),fx=0.75,fy=0.75)
plt.imshow(cv2.cvtColor(newImg, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
Img_change_HW = cv2.resize(img, (300,200))
plt.imshow(cv2.cvtColor(Img_change_HW, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
y1 = 100
x1 = 100
y2 = 300
x2 = 200

cropped_img = img[y1:y2,x1:x2]
print("Cropped Image \n")
plt.imshow(cv2.cvtColor(cropped_img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python

#RGB
imgRGB = cv2.cvtColor(img,cv2.COLOR_RGB2BGR)
cv2.imshow('image',imgRGB)
cv2.waitKey(0)

#HSV
imgHSV = cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
cv2.imshow('HSV',imgHSV)
cv2.waitKey(0)

#HLS
imgHLS = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
cv2.imshow('HLS',imgHLS)
cv2.waitKey(0)

#gray
imgGRAY = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow('GRAY',imgGRAY)
cv2.waitKey(0)
```

```python
img = cv2.imread('assets/images/gunung.jpg')
print("Image 1")
plt.imshow(cv2.cvtColor(img, cv2.COLOR_RGB2BGR))
plt.show()

img2 = cv2.imread('assets/images/semangka.jpg')
print("Image 2")
plt.imshow(cv2.cvtColor(img2,cv2.COLOR_BGR2RGB))
plt.show()

img = cv2.resize(img,(750,750))
img2 = cv2.resize(img,(750,750))

blended = cv2.addWeighted(src1=img,alpha=0.6,src2=img2,beta=0.1,gamma=0)
plt.imshow(cv2.cvtColor(blended, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img = cv2.imread('assets/images/semangka.jpg')
font = cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img, text='GREAT PICTURE', org=(10,300), fontFace=font, fontScale=4, color=(0,255,0), thickness=4)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```

```python
blurred = cv2.GaussianBlur(img,(7,7),0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
```

```python
img = cv2.imread('assets/images/semangka.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
plt.imshow(cv2.cvtColor(thresh1,cv2.COLOR_BGR2RGB))
```

```python

ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
plt.imshow(cv2.cvtColor(thresh2,cv2.COLOR_BGR2RGB))
```

```python
img = cv2.imread('assets/images/semangka.jpg')

edges = cv2.Canny(img,100,200)
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
```
